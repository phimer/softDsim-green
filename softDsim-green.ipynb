```python
# Import seaborn
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
co2 = {'no compression': [0.355, 0.354,0.355,0.355,0.355],
    'brotli': [0.264,0.264,0.264,0.264,0.264],
    'gzip': [0.265,0.265,0.266,0.265,0.265]}

df = pd.DataFrame(data=co2)
df.index = np.arange(1, len(df) + 1)
df
```

```python
mhw = {'no compression': [0.804, 0.801,0.803,0.803,0.803],
    'brotli': [0.597,0.597,0.597,0.597,0.598],
    'gzip': [0.599,0.599,0.601,0.600,0.601]}

df2 = pd.DataFrame(data=mhw)
df2.index = np.arange(1, len(df) + 1)
df2
```

```python

# Apply the default theme
sns.set_theme(style="white")

fig, axes = plt.subplots(1,2, figsize=(12, 4))

# plot 1
ax1 = sns.lineplot(
    data=df,
    markers=True,
    ax=axes[0]
)
ax1.set(xlabel='Test-Run', ylabel='CO2 emissions (grams) per execution ')
ax1.set(ylim=(0.25, 0.4))
ax1.set(xticks=[1,2,3,4,5])
ax1.set(title='Runs of Experiment 2')

# plot 2
ax2 = sns.lineplot(
    data=df2,
    markers=True,
    ax=axes[1]
)
ax2.set(xlabel='Test-Run', ylabel='Electricity usage (Wh) per execution ')
ax2.set(ylim=(0.5, 1))
ax2.set(xticks=[1,2,3,4,5])
ax2.set(title='Runs of Experiment 2')

plt.show()
```
